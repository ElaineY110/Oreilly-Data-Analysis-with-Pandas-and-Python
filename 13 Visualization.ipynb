## Create Bar Graph to Show Counts
plot(kind = 'bar'/'barh')

```python
samsung.Close
```

```python
def rank_performance(stock_price):
    if stock_price <= 50000:
        return 'Poor'
    elif stock_price <= 60000:
        return 'Satisfactory'
    else:
        return 'Stellar'
```

```python
samsung.Close.apply(rank_performance).value_counts()
```

```python
samsung.Close.apply(rank_performance).value_counts().plot(kind = 'bar')
```

```python
samsung.Close.apply(rank_performance).value_counts().plot(kind = 'barh')
```

## Creating Pie Charts to Represent Proportions

```python
samsung.Close.mean()
```

```python
samsung.describe()
```

```python
def rank_performance(stock_price):
    if stock_price >= 50360:
        return 'Above Average'
    else:
        return 'Below Average'
```

```python
plt.style.use('dark_background')
samsung.Close.apply(rank_performance).value_counts().plot(kind = 'pie', legend = True)
```
